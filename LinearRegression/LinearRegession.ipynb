```python
from __future__ import division, print_function, unicode_literals
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

## Đọc dữ liệu từ file csv
## Thông tin bài toán:
Dự đoán lương của một nhân viên dựa trên số năm kinh nghiệm của họ

```python
df = pd.read_csv('Salary_Data.csv')
```

```python
print(df.head())
```

```python
X = df[['YearsExperience']]
y= df[['Salary']]
```

## Biểu diễn dữ liệu

```python
df.plot(x='YearsExperience', y='Salary', kind='scatter', marker='o', title='Salary vs. Years of Experience Chart:')
plt.ylabel('Salary: USD')
plt.show()
```

Sử dụng linear regresion để vẽ đường dự đoán.

Nghiệm tối ưu cho bài toán:
$$\mathbf{w} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}$$

```python

one = np.ones((X.shape[0], 1))
Xbar = np.concatenate((one, X), axis = 1)
A = np.dot(Xbar.T, Xbar)
b = np.dot(Xbar.T, y)
w = np.dot(np.linalg.pinv(A), b)
print('w = ', w)
w_0 = w[0][0]
w_1 = w[1][0]
x0 = np.linspace(0, 12, 2)
y0 = w_0 + w_1*x0
plt.plot(X, y, 'ro')  
plt.plot(x0, y0)           
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

```python
y1 = w_1*5.5 + w_0
y2 = w_1*1.5 + w_0
print( 'Predict salary of person with 5.5 YearsExperience: %.2f (USD),'  %(y1) )
print( 'Predict salary of person with 1.5 YearsExperience: %.2f (USD),'  %(y2) )
```
